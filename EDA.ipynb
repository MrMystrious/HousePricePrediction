```python
from bivariateAnalysis import NumericalVsNumericalAnalysis,CategoricalVsCategoricalAnalysis,CategoricalVsNumericalAnalysis,BivariateStrategy
```

```python
from multivariateAnalysis import SimpleMultivariateAnalysis
```

```python
ObjectColumns = df.select_dtypes(include=["object"]).columns
IntColumns = df.select_dtypes(include=["int64"]).columns
FloatColumns = df.select_dtypes(include=["float64"]).columns
bivariateAnalyser = BivariateStrategy(NumericalVsNumericalAnalysis())
target = "SalePrice"
bivariateAnalyser.set_strategy(CategoricalVsNumericalAnalysis())

anovavalues = {}
AffectingFeatures = [target]
for category in ObjectColumns:
    
    anovavalues[category] = bivariateAnalyser.Anova(df,category,target,True)

for key in anovavalues.keys():
    if(anovavalues[key][1] < 1e-50):
        AffectingFeatures.append(key)
        print(key,anovavalues[key])    
```

```python
bivariateAnalyser.set_strategy(NumericalVsNumericalAnalysis())
anovavaluesNumerical={}

for category in IntColumns:
    if(category == target):
        continue
    anovavaluesNumerical[category] = bivariateAnalyser.Anova(df,target,category,True)

for key in anovavaluesNumerical.keys():
    if(anovavaluesNumerical[key][1] < 1e-50):
        AffectingFeatures.append(key)
        print(key,anovavaluesNumerical[key])  
```

```python
anovaFloating = {}
for category in FloatColumns:
    if(category == target):
        continue
    anovaFloating[category] = bivariateAnalyser.Anova(df,target,category,True)

for key in anovaFloating.keys():
    if(anovaFloating[key][1] < 1e-50):
        AffectingFeatures.append(key)
        print(key,anovaFloating[key])
    
anovaFloating
```
